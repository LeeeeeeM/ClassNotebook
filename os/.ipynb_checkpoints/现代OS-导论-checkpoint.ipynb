### 进程 
Process，本质上是一个正在执行的程序。
* 与进程息息相关的概念有地址空间(address space)，里面存放着进程的内容
* shell进程是一个特殊的进程
* 合作作业通常会构成一颗进程树，其中树上一个进程能够创建多个进程，称为其子进程。
* 进程之间的通信成为进程间通信
* 在一些非常简单的操作系统中，内存中一次只能有一个程序，如果要运行第二个程序，那么第一个程序就必须移出内存
* OS必须为进程分配合适的地址空间
* 有一些程序需要大量的被分配的地址空间，甚至超过了硬件所能容纳的大小，这时候OS的虚拟化技术就非常有用了（大部分的内容可能都留在disk上），虚拟化了虚拟内存作为物理内存的抽象。

### 文件系统
OS提供一个良好的文件系统抽象，对于文件的许多操作都需要系统调用。
* 进程和文件层次都可组织为树状结构
* 每个进程都有一个工作目录
* 管道是一种虚文件，可连接两个进程，进行数据的传输。

### shell
操作系统是进行系统调用的代码。编辑器，编译器，汇编程序，链接程序以及命令解释器尽管非常重要，但确实不是操作系统的组成部分。UNIX的命令解释器就是shell。
* 注意shell并不是操作系统的一部分
* shell是终端的用户与操作系统之间的界面

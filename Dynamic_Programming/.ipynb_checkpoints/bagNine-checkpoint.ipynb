```python
def DpSolve(valuedict:dict, volume):
    """
    valuedict是装每件物品对应的消耗以及价值的容器:
    元组的第一个元素是消耗，第二个元素是价值
    volume是背包的容量
    
    valuedict[i][0]
    """
    F = {item:[0] * (volume + 1) for item in range(0, len(valuedict)+1)}
    for i in range(1, len(valuedict)+1):
        for v in range(1, volume + 1):
            if valuedict[i][0] <= v:
                F[i][v] = max(F[i-1][v], F[i-1][v-valuedict[i][0]] + valuedict[i][1])
            else:
                F[i][v] = F[i-1][v]
    return F
```

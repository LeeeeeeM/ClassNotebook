

CPU由两部分组成
* Control Unit: 负责控制。包括计数器，指令寄存。
* Logic Unit: 负责运算。加法器，累加器。

CPU中还有存储单元等其他单元。
* CPU并没有驱动大电流的能力，苦力活（如操纵驱动电路）交给专业的设备（如显卡对图像处理）去做
* CPU中的核心share存储单元



> 特斯拉自动驾驶芯片。内部结构包含两个NPU计算单元，就是再计算神经网络。

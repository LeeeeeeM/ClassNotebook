## 从电开始
虽然被静电电到的时候我们是看不见的，但会感觉电是在导体上游走的微小颗粒。18实际，科学家们为这种微小的粒子起名为电子。

回顾一下原子的知识，原子里有原子核，原子核里面由质子，原子核外有电子，决定原子种类的是质子的数量。物质的形成机制是原子共用外层电子，导体的共用电子通常不稳定，他们被称为自由电子，当你用墙上插座里的点自来推动导体自由电子朝一个方向移动时，“导电”就发生了。

当电子朝一个方向前进时，形成了电流。实际上，感觉电流速度是“一瞬间”，或者说认为他不需要“速度”进行衡量，这是一种普遍的错觉。

说到底，电池或者发电厂做的并不是制造电子，而是让导体中的电子循环流动，也就是说，“电泵”们从电线一端夺走自由电子并将他们送给电线的另一端，这种时候电压就产生了。

要形成电路，电源，导线和用电器件都是必需品。

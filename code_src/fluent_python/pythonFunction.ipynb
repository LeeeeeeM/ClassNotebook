```python
# 扩展话题，尾递归
def factorial(n):
    '''return n!'''
    return 1 if n < 2 else n * factorial(n-1)
```

```python
fact = factorial
fact
```

```python
list(map(fact, range(10)))
```

```python
# 可调用对象的建立
import random

class BingoCage:
    def __init__(self, items):
        #接受任何可迭代的对象
        self._items = list(items)
        random.shuffle(self._items)
        
    def pick(self):
        try:
            return self._items.pop()
        except IndexErrore:
            raise LookupError('pick from empty BingoCage')
            
    def __call__(self):
        return self.pick()
```

```python
bingo = BingoCage(range(3))
```

```python
def calll(obj,times:int):
    for i in range(times):
        obj()
```

```python
calll(bingo, 3)
```

```python
bingo()
```

```python
initState = [True] * 1000
def ChangeState(obj):
    for i in range(1,len(obj)+1):
        tem = i
        while tem <= len(obj):
            obj[tem-1] = not obj[tem-1]
            tem += i

ChangeState(initState)
initState
```
